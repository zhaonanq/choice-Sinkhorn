```python
matrix = matrix[:50000,:]
```

```python
import networkx as nx
```

```python
matrix.shape
```

```python
G = nx.DiGraph()
```

```python
#build ranking data, need to select largest strongly connected component
rankings = list(map(tuple, matrix))
```

```python
rankings
```

```python
G.add_edges_from(rankings)
```

```python
G = max(nx.strongly_connected_component_subgraphs(G), key=len)
```

```python
Adjacency = nx.adjacency_matrix(G)
```

```python
Adjacency = Adjacency.toarray()
```

```python
p = np.sum(Adjacency,1)
```

```python
n = np.sum(Adjacency)
```

```python
m = Adjacency.shape[0]
```

```python
A = np.zeros((n,m))
```

```python
#np.squeeze((Adjacency[1,:]==1).toarray())
```

```python
index = 0
for i in range(m):
    length = np.sum(Adjacency[i,:])
    pair = np.where(Adjacency[i,:]==1)[0]
    A[index:index+length,i] = 1
    A[np.arange(index,index+length),pair] = 1
    index += length
    print(A)
```
